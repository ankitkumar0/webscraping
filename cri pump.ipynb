```python
city="/html/body/div[1]/section[2]/section[3]/div/form/span[2]/select/option[{}]"
comp="/html/body/div[1]/section[2]/section[3]/div/div/div[2]/span[{}]"
num="/html/body/div[1]/section[2]/section[3]/div/div/div[2]/span[{}]"

company=[]
number=[]
for i in range(2,45):
    dd=city.format(i)
    try:
        driver.find_element_by_xpath(dd).click()
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="right-side"]/section[3]/div/form/input').click()
        time.sleep(2)
    except:
        pass
    for i in np.arange(1,500,5):
        ank=comp.format(i)
        try:
            r=driver.find_element_by_xpath(ank).text
            company.append(r)
        except:
            pass
    for i in np.arange(3,500,5):
        ank1=num.format(i)
        try:
            r=driver.find_element_by_xpath(ank1).text
            number.append(r)
        except:
            pass
    
        
```
