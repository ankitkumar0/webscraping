```python
def get_url(search_term):
    """"genrate a url from search term"""
    template="https://www.amazon.in/s?k={}&ref=nb_sb_noss_2"
    search_term=search_term.replace(" ","+")
    url=template.format(search_term)
    url+='&page{}'

    return url
    
    
def extract_record(item):
    atag=item.h2.a
    discription=atag.text.strip()
    url="https://www.amazon.in"+ atag.get('href')
    try:
        price_parent=item.find('span',{'class':'a-price'})
        price=price_parent.find('span','a-offscreen').text
    except AttributeError:
        return
    try:
        rating=item.i.text
        review_count=item.find('span',{'class':"a-size-base",'dir':"auto"}).text
    except AttributeError:
        rating=" "
        review_count=" "
        
        
    result=(discription,url,price,rating,review_count)
    return result


def main(search_term):
    driver = webdriver.Chrome("C:\\Users\\Deepak Gupta\\Downloads\\chromedriver")
    
    records=[]
    url=get_url(search_term)
    
    for page in range (1,3):
        driver.get(url.format(page))
        soup=BeautifulSoup(driver.page_source,'html.parser')
        results=soup.find_all('div',{"data-component-type":"s-search-result"})
        
        for item in results:
            record=extract_record(item)
            if record:
                records.append(record)
            
    driver.close()
    
    #records.to_csv('CarpetsRugs.csv', mode='a', header=True)
    
    with open('results.csv','w',newline='',encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['discription','url','price','rating','review_count'])
        writer.writerows(records)
    
        
                   
```
